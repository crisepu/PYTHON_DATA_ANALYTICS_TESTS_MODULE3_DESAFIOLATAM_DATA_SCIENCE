**1.f.** Generamos una tabla pivote que muestre el número de casos por mes en la fila y por año en las columnas.

En primer lugar generamos dos columnas nuevas que solo contengan el año y el mes respectivamente

```python
df_merged['date'] = pd.to_datetime(df_merged['date'])
df_merged['year'] = df_merged['date'].dt.year
df_merged['month'] = df_merged['date'].dt.month
df_merged.head()
```

Procedemos a generar la tabla pivote solicitada

```python
df_pivot1 = pd.pivot_table(
                            data = df_merged,
                            index = 'month',
                            columns = 'year',
                            values = 'case_number',
                            aggfunc = 'count',
                            margins = True
)

df_pivot1
```

Con la tabla es posible identificar meses y años en los que suelen haber un mayot numero de casos, por ejemplo, junio ha sido uno de los meses en los que mas se ha generado casos, en especial en el año 2015 en donde se registraron 18 casos solo en ese mes. Y es el mismo año 2015 que en su totalidad muestra un mayor numero de csos alcanzando los 40.

**1.g.** Generamos una nueva tabla pivote que muestre en las filas el género del oficial y en las columnas el género del subject.

```python
df_pivot2 = pd.pivot_table(
                            data = df_merged,
                            index = 'gender_off',
                            columns = 'gender_sub',
                            values = 'case_number',
                            aggfunc = 'count',
                            margins = True
)

df_pivot2
```

**2.a.** Cargamos los datos y generamos un dataframe con el archivo Cleaned_DS_Jobs.csv

```python
df_cleaned = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Cleaned_DS_Jobs.csv')
df_cleaned.head()
```

**2.b.** Procedemos a reemplazar los valores que seran considerados como nulos

```python
null_values = ["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"]
df_cleaned = df_cleaned.replace(to_replace = null_values, value = np.nan)
df_cleaned.head()
```

**2.c.** Eliminamos todas las filas con datos faltandes **(Nan no es dato faltante)**

```python
df_cleaned.dropna(axis = 0)
df_cleaned.reset_index
df_cleaned.head()
```

**2.d.** A partir de la columna “Salary Estimate”, generamos dos columnas: Salario Estimado Mínimo y Máximo.

---



```python
df_cleaned['salario_estimado_minimo'] = df_cleaned['Salary Estimate'].apply(lambda x: int(x.split("-")[0])) #Parseamos a numerico con int() para no tener problemas con el calculo de promedio a futuro.
df_cleaned['salario_estimado_maximo'] = df_cleaned['Salary Estimate'].apply(lambda x: int(x.split("-")[1]))
df_cleaned.head()
```

**2.e** Realizamos la recodificación de la columna Size con los valores de la siguiente tabla

```python
#Para esto generaremos inicialmente un diccionario y eemplazaremos con la funcion replace()
replace_dict = {'10000+ employees': 'Mega Empresas', '5001 to 10000 employees': 'Grandes Empresas', '1001 to 5000 employees': 'Medianas Empresas', '501 to 1000 employees': 'Pequenas Grandes Empresas', '201 to 500 employees': 'Pequenas Empresas', '51 to 200 employees': 'Microempresas', '1 to 50 employees': 'Pymes', 'Unknown': 'Empresas sin informacion'}
df_cleaned_replaced = df_cleaned.replace(replace_dict)
df_cleaned_replaced.head()
```

**2.f.** Finalmente, generaremos una tabla pivote que muestre la media del salario estimado mínimo y la media del salario estimado máximo por tamaño de empresa

```python
df_pivot3 = pd.pivot_table(
                            data = df_cleaned_replaced,
                            index = 'Size',
                            values = ['salario_estimado_minimo', 'salario_estimado_maximo'],
                            aggfunc = {'salario_estimado_minimo': "mean", 'salario_estimado_maximo': "mean"},
                            margins = True
)

df_pivot3
```
