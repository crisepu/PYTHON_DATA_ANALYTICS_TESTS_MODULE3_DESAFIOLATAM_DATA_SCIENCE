**2. Fabricacion de cerveza artesanal**

Importamos la libreria NumPy

```python
import numpy as np
```

Leemos el archivo. **Para la carga de los datos, se esta alojando el archivo alcohol_cerveza.txt en la carpeta Colab Notebooks de google drive, para efectos de revision, se sugiere al revisor alojar el documento en la misma ubicacion o modificar la ubicacion para la carga.**

```python
# necesario para montar documento desde google drive
from google.colab import drive

#se solicitará acceso a su google drive
drive.mount('/content/drive')
array = np.loadtxt('/content/drive/MyDrive/Colab Notebooks/alcohol_cerveza.txt')
print(array)
```

Calculamos la media de los datos

```python
media = np.mean(array)
print(media)
```

Calculamos la desviacion estandar de los datos

```python
desviacion = np.std(array)
print(desviacion)
```

Calculamos el Limite Inferior de Control (LIC)

```python
lic = media - (3 * desviacion)
print(lic)
```

Calculamos el Limite Superior de Control

```python
lsc = media + (3 * desviacion)
print(lsc)
```

En resumen, los calores calculados son los siguientes:

```python
print(f'La media de los datos es: {media}')
print(f'La desviacion estandar de los datos es: {desviacion}')
print(f'El limite inferior de control (LIC) es: {lic}')
print(f'El limite superior de control (LSC) es: {lsc}')
```

Finalmente filtramos e imprimimos los valores fuera del rango de control

```python
filter_array = array[np.logical_or(array < lic, array > lsc)]
print(filter_array)
```
