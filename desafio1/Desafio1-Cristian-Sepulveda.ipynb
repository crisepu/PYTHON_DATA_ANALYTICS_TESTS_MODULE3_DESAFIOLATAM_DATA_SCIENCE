```python
# 1. Calculando el salario semanal

# se define el salario base
salarioBase = 500

""" se crea un bucle while que permitar ingresar y guardar las ventas totales
y a la vez validar si es un valor valido el ingresado. En caso de no ser
valido, indica cual es el error y permite ingresar el valor nuevamente """
while True:
    try:
        ventasTotales = int(input('Ingrese ventas totales: '))
    except ValueError:
        print('Debes escribir un número.')
        continue

    if ventasTotales < 0:
        print('Debe ingresar un valor mayor o igual a cero.')
        continue
    else:
        break

# se hace el calculo de salario total
salarioTotal = salarioBase + ventasTotales*0.1

#se muestra por pantalla el resultado final
print(f'Su salario total es de: {salarioTotal}')
```

```python
# 2. Bienvenido a la fruteria virtual

#se geneta el diccionario con las frutas y sus valores
catalogo_frutas = {'manzana': 1000,
                   'pera': 1200,
                   'platano': 1400,
                   'kiwi': 1500,
                   'naranja': 1000,
                   'mandarina': 800,
                   'frutilla': 3000
                   }

#se guarda el valor ingresado por el usuario
nombre_fruta = input('Ingrese nombre de la fruta: ')

#se pasa todo a minusculas por si el usuario ingresa nombre con mayuscula
nombre_fruta = nombre_fruta.lower()

"""se verifica si la fruta esta presente en el diccionario, en caso de estar
presente, se indica al usuario el precio, en caso contrario, se indica que la
fruta no se encuentra en el catalogo"""
if nombre_fruta in catalogo_frutas:
  print(f'El valor de {nombre_fruta} es de ${catalogo_frutas[nombre_fruta]} el kg.')
else:
  print(f'LA fruta {nombre_fruta} no se encuentra en la tienda')
```

```python
# 3. Bienvenido al casino

#importamos la libreria random
import random

"""se le solicita el nombre mas que nada para tener un punto de inicio y hacer
el juego mas personalizado"""
nombre = input('Por favor ingrese un nombre para iniciar: ')

#se generaon los dados con numeros aleatorios de 1 a 6 inclusive
dado_1 = random.randint(1,6)
dado_2 = random.randint(1,6)

#se suma el valor de los dados
suma_dados = dado_1 + dado_2

#Se se muestran los valores al usuario
print(f'{nombre}!, lo resultados de este emocionante juego de dados son los siguientes:')
print(f'El resultado del primer dado es {dado_1}')
print(f'El resultado del segundo dado es {dado_2}')
print(f'La suma de ambos dados es {suma_dados}')
```

```python
# 4. Caculadora de chocolate

""" se crea un bucle while que permita ingresar y guardar el limite de
calorias diarias y a la vez validar si es un valor valido el ingresado.
En caso de no ser valido, indica cual es el error y permite ingresar el
valor nuevamente """
while True:
    try:
        limite_diario_calorias = int(input('Ingrese su limite diario de calorias: '))
    except ValueError:
        print('Debes escribir un número.')
        continue

    if limite_diario_calorias <= 0:
        print('Debe ingresar un valor mayor a cero.')
        continue
    else:
        break

""" se crea un bucle while que permita ingresar y guardar el la cantidad
de calorias por porcion de chocolate y a la vez validar si es un valor
valido el ingresado. En caso de no ser valido, indica cual es el error y
permite ingresar el valor nuevamente """
while True:
    try:
        calorias_porcion = int(input('Ingrese la cantidad de calorias por porcion de chocolate: '))
    except ValueError:
        print('Debes escribir un número.')
        continue

    if calorias_porcion <= 0:
        print('Debe ingresar un valor mayor a cero.')
        continue
    else:
        break

#se hace el calculo se las porciones de chocolate maximas a consumir
cantidad_max_chocolate = int(limite_diario_calorias / calorias_porcion)

#se entrega al usuario las cantidad maxima de porciones diarias de chocolate
print(f'La cantidad de porciones que puedes comer diariamente de ese chocolate es {cantidad_max_chocolate}')
```
