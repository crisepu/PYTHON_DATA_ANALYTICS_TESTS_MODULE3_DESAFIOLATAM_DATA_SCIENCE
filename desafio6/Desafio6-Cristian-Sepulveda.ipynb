**2.** Hacemos un cruce de todos los dataframes usando la validacion many_to_one

```python
#Primero mergeamos los dataframes sales_orders y customers
df_base = df_sales_orders.merge(df_customers, on='_CustomerID', validate = 'many_to_one')
df_base.head()
```

```python
#Al dataframe resultante, le mergeamos el dataframe store_locations
df_base = df_base.merge(df_store_locations, on='_StoreID', validate = 'many_to_one')
df_base.head()
```

```python
#Al dataframe resultante, le mergeamos el dataframe products
df_base = df_base.merge(df_products, on='_ProductID', validate = 'many_to_one')
df_base.head()
```

```python
#Al dataframe resultante, le mergeamos el dataframe products
df_base = df_base.merge(df_sales_team, on='_SalesTeamID', validate = 'many_to_one')
df_base.head()
```

**3.** Procedemos a generar las columnas indicadas, asegurandonos de que las columnas sean de tipo datetime

```python
#En primer lugar nos aseguraremos que todas las columnas involucradas en el calculo de las nuevas columnas sean del tipo datetime
df_base['DeliveryDate'] = pd.to_datetime(df_base['DeliveryDate'])
df_base['ProcuredDate'] = pd.to_datetime(df_base['ProcuredDate'])
df_base['ShipDate'] = pd.to_datetime(df_base['ShipDate'])
df_base['OrderDate'] = pd.to_datetime(df_base['OrderDate'])
df_base['DeliveryDate'] = pd.to_datetime(df_base['DeliveryDate'])
df_base['DeliveryDate'] = pd.to_datetime(df_base['DeliveryDate'])
```

```python
#Procedemos a crear las nuevas columnas
df_base['ProcurementDays'] = df_base['OrderDate'] - df_base['ProcuredDate']
df_base['ShippingDays'] = df_base['ShipDate'] - df_base['OrderDate']
df_base['DeliveryDays'] = df_base['DeliveryDate'] - df_base['ShipDate']
df_base['CustomerDays'] = df_base['ShippingDays'] - df_base['DeliveryDays']
df_base.head()
```

**4.** Procedemos a generar la columna discreta segun lo solicitado, pero antes, procederemos a hacer algunos ajuster a la columna CustomerDays

```python
#Primero dejaremos CustomerDays solo con su valor numerico
df_base['CustomerDays'] = df_base['CustomerDays'].dt.days
df_base.head()
```

```python
#Para efectos de la generacion de intervalores, los valores negativos los dejaremos como 0, para estaremos usaremos la funcion where de numpy.
df_base['CustomerDays'] = np.where(df_base['CustomerDays'] < 0, 0, df_base['CustomerDays'])
df_base.head()
```

```python
#Hecho lo anterior, procedemos a generar los intervalos
intervals = [-1, 15, 30, 45, 60, 75, 90]
labels = ['0 to 15 days',
          '15 to 30 days',
          '30 to 45 days',
          '45 to 60 days',
          '60 to 75 days',
          '75 to 90 days']

df_base['CustomerDaysInterval'] = pd.cut(df_base['CustomerDays'], bins=intervals, labels=labels)
df_base.head()
```

**5.** Generamos una tabla pivote para mostrar el conteo de órdenes agrupando por Sales Team en las filas y CustomerDaysInterval en las columnas.

```python
df_pivot1 = pd.pivot_table(
                            data = df_base,
                            index = 'Sales Team',
                            columns= 'CustomerDaysInterval',
                            values = 'OrderNumber',
                            aggfunc = 'count',
                            margins = True
)

df_pivot1
```

**6.** Agregamos la columna GrossMargin

```python
df_base['GrossMargin'] = df_base['Order Quantity'] * (df_base['Unit Price'] * (1 - df_base['Discount Applied']) - df_base['Unit Cost'])
df_base.head()
```

**7.** Procedemos a crear la columna CommisionsPercentage

```python
intervals = [-1, 100, 1000, 10000, 100000]
labels = [0.05,
          0.1,
          0.15,
          0.2]

df_base['CommissionsPercentage'] = pd.cut(df_base['GrossMargin'], bins=intervals, labels=labels)
df_base.head()
```

Usaremos la funcion .isnull() para cehquear si hay algun porcentage de comision fuera de rango (NaN)

```python
df_base['CommissionsPercentage'].isnull()
```

Ya que existen columnas NaN en ComissionsPercentage, resultado de que no se encuentran dentro de los rangos definidos, usaremos la funcion where de numpy para convertir estos valores en 0, de esta manera no tendremos problemas en el siguiente calculo de la columna CommissionsAmount.

```python
df_base['CommissionsPercentage'] = np.where(df_base['CommissionsPercentage'].isnull(), 0, df_base['CommissionsPercentage'])
```

**8.** Procedemos a calcular el monto de la comision en la columna CommissionsAmount

```python

df_base['CommissionsAmount'] = df_base['GrossMargin'] * df_base['CommissionsPercentage']
df_base.head()
```

**9.** Calculamos la comision sobre el margen bruto en la columna NetMargin

```python
df_base['NetMargin'] = df_base['GrossMargin'] - df_base['CommissionsAmount']
df_base.head()
```

**10.** Finalmente generamos una nueva tabla pivote mostrando la suma de los valores GrossMargin, NetMargin y CommissionAmount agrupando por Sales Team en las filas

```python
df_pivot2 = pd.pivot_table(
                            data = df_base,
                            index = 'Sales Team',
                            values = ['GrossMargin', 'NetMargin', 'CommissionsAmount'],
                            aggfunc = {'GrossMargin': "sum", 'NetMargin': "sum", 'CommissionsAmount': "sum"},
                            margins = True
)

df_pivot2
```
