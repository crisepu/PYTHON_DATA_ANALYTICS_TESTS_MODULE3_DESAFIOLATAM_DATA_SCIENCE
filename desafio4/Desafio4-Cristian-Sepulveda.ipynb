**1.** Creamos el dataframe con el archivo .csv entregado

```python
import pandas as pd

#Ajustar la ruta de acuerdo a la donde ubique el archivo
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/resumen_resultados_astronautas.csv')
#Hacemos un head para verificar que el dataframe haya sido creado correctamente
df.head()
```

**2.** Creamos la columna "IMC" usando la formula IMC = m / h*h

```python
df['IMC'] = df['peso'] / (df['altura']/100)**2
df.head()
```

**3.a** Creamos la columna "promedio_evaluaciones" con el promedio de las 4 evaluaciones

```python
df['promedio_evaluaciones'] = (df['evaluacion_fisica'] + df['evaluacion_psicologica'] + df['evaluacion_adaptativa'] + df['evaluacion_medica']) / 4
df.head()
```

**3.b** Hacemos un filtro para extraer las filas cuyo IMC está entre 18.5 y 24.9

```python
df_imc = df.query("IMC >= 18.5 and IMC <= 24.9")
df_imc.head()
```

**4.** Hacemos un filtro para extraer las filas cuyo promedio_evaluaciones sea mayor o igual a 87

```python
df_prom_eval = df_imc.query("promedio_evaluaciones >= 87")
df_prom_eval
```

**5.** Reiniciamos los indices de este ultimo dataframe

```python
df_result = df_prom_eval.reset_index(drop=True)
df_result
```

**6.** Modificamos el valor de de la columna califica a si en todas las filas del dataframe

```python
df_result['califica'] = df_result['califica'].replace(['No'], 'Si')
df_result
```

**7.** Exporta el DataFrame resultante a un archivo .csv separado por comas con el nombre "astronautas_calificados.csv"

```python
df_result.to_csv("astronautas_calificados.csv", sep = ',', header = False)
```
