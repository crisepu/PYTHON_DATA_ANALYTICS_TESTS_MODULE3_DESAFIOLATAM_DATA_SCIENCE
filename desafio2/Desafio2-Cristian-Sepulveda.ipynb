## 1. Ejercicio 1: Suma de numeros pares

En primer lugar pedimos al usuario dos numeros enteros, pero esto usamos un bucle while junto al metodo try de manera de asegurarnos que el usuario ingrese los numeros con las caracteristicas esperadas. Si el usuario no ingresa un numero entero, este sera solicitado nuevamente:

```python
while True:
    try:
        number_1 = int(input('Ingrese el primer numero entero: '))
    except ValueError:
        print('El primer numero ingresado debe ser un numero entero')
        continue
    break

while True:
    try:
        number_2 = int(input('Ingrese el segundo numero entero: '))
    except ValueError:
        print('El segundo numero ingresado debe ser un numero entero')
        continue
    break
```

## 2. Ejercicio 2: Calculadora de areas

Primero que todo, solicitamos al usuario que ingrese una opcion de figuta valida, para este se le pedira que ingrese un numero como opcion. Con el bucle while y la funcion try verificaremos que el usuario ingrese una opcion valida:

```python
while True:

  print('Las opciones validas para el calculo del area son: ')
  print('Cuadrado  : 1')
  print('Triangulo : 2')
  print('Circulo   : 3')

  try:
        option = int(input('Ingrese una opcion para el calculo del area: '))
  except ValueError:
    print('La opcion ingresada no es valida.')
    continue

  if option < 1 or option > 3:
    print('La opcion ingresada no es valida.')
    continue
  else:
    break
```

Procedemos a definir funciones para el calculo de las distintas areas de esta forma los siguientes pasos quedaran mas ordenados.

```python
import math

#funcion para el calculo de un cuadrado
def areaCuadrado(lado):
  return lado * 4

#funcion para el calculo de un triangulo
def areaTriangulo(base, altura):
  return (base * altura) / 2

#funcion para el calculo de un circulo
def areaCirculo(radio):
  return math.pi * radio^2
```

Agregamos una funcion adicional para la captura y validacion de los valores ingresados por el usuario (estos no pueden ser negativos):

```python
def input_validator(dimension) :
  while True:
    try:
        value = float(input(f'{dimension}: '))
    except ValueError:
        print('Debes escribir un número.')
        continue

    if value <= 0:
        print('Debe ingresar un valor mayor a cero.')
        continue
    else:
        break

  return value
```

Hecho lo anteiror, finalizamos juntad todo y utilizar la opcion ingresada por el usuario para realizar el calculo e imprimir por pantalla:

```python
if option == 1:
  lado = input_validator('lado')
  result = areaCuadrado(lado)
  print(f'El area del cuadrado de lado: {lado} es {result} unidades cuadradas.')
elif option == 2:
  base = input_validator('base')
  altura = input_validator('altura')
  result = areaTriangulo(base, altura)
  print(f'El area del triangulo de base: {base} y altura: {altura} es {result} unidades cuadradas.')
else:
  radio = input_validator('radio')
  result = areaCirculo(radio)
  print(f'El area del circulo de radio: {radio} es {result} unidades cuadradas.')
```

```python
while True:

  print('Estas son las velocidades disponibles (en escala caracol) ')
  print('Alta velocidad        : 1')
  print('Velocidad supersonica : 2')
  print('Velocidad luz         : 3')

  try:
        speed = int(input('Ingrese la velocidad a la que quiere que corra el caracol: '))
  except ValueError:
    print('La opcion ingresada no es valida.')
    continue

  if speed < 1 or speed > 3:
    print('La opcion ingresada no es valida.')
    continue
  else:
    break
```
